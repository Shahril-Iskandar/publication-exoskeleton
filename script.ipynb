# Data cleaning

```python
# Function to filter dataset
def filter_dataset(df, equipment, mass, incline_condition):
    index = (df[1]==equipment)&(df[2]==mass)&(df[3]==incline_condition)
    df2 = df[index]
    df2 = df2.sort_values(df2.columns[0], ascending=True)
    print(f"{equipment} {mass} {incline_condition}: {df2[0].nunique()} unique rows")
    df2 = df2.iloc[:,4:]
    print(f"{equipment} {mass} {incline_condition}: {df2.shape[1]} columns\n")
    return df2
```

# SPM analysis

```python
def spm_graph(non_exo_flat, exo_flat,
              non_exo_incline, exo_incline,
              non_exo_decline, exo_decline,
              leg_position: str, load: str):
    #(1) Perform SPM:
    spm1 = spm1d.stats.ttest_paired(non_exo_flat.to_numpy(), exo_flat.to_numpy())
    spmi1 = spm1.inference(0.05, two_tailed=True, interp=True)

    spm2 = spm1d.stats.ttest_paired(non_exo_incline.to_numpy(), exo_incline.to_numpy())
    spmi2 = spm2.inference(0.05, two_tailed=True, interp=True)

    spm3 = spm1d.stats.ttest_paired(non_exo_decline.to_numpy(), exo_decline.to_numpy())
    spmi3 = spm3.inference(0.05, two_tailed=True, interp=True)
    # print(spmi)

    #(2) Plot:
    # plt.close('all')
    # fig, AX = plt.subplots(2, 3, figsize=(15, 12) )
    fig = plt.figure(figsize=(15,10))
    gs = gridspec.GridSpec(nrows=2, ncols=3, height_ratios=[1,0.65])

    ### Flat
    # plot mean and SD:
    condition = "Flat"
    title_flat = f"{leg_position} - {load} {condition}"
    ax = fig.add_subplot(gs[0,0])
    # ax = AX[0, 0] # First row
    ax.text(-0.1, 1.1, "1a", transform=ax.transAxes, size=18, weight='bold')
    ax.set_title(title_flat, fontsize=24)
    plt.sca(ax)
    spm1d.plot.plot_mean_sd(exo_flat, linecolor='r', facecolor='r', label='Exo')
    spm1d.plot.plot_mean_sd(non_exo_flat, linecolor='b', facecolor='b', label='Non-Exo')
    ax.axhline(y=0, color='k', linestyle=':')
    # ax.set_xlabel('Stance phase (%)')
    ax.set_ylabel('Normalised GRF (BW)', fontsize=20)
    ax.set_xlabel('Stance phase (%)', fontsize=20)
    ax.set_ylim([0, 2.7])
    ax.set_xlim([0, 100])
    plt.xticks(fontsize=17)
    plt.yticks(fontsize=17)
    ax.legend(fontsize=17)

    # plot SPM results:
    # ax = AX[1, 0]
    ax = fig.add_subplot(gs[1,0])
    ax.text(-0.1, 1.1, "1b", transform=ax.transAxes, size=18, weight='bold')
    plt.sca(ax)
    spmi1.plot()
    spmi1.plot_threshold_label(fontsize=17, pos=(20,5.5))
    spmi1.plot_p_values(size=10, offsets=[(0,0.3)])
    ax.set_xlabel('Stance phase (%)', fontsize=20)
    ax.set_ylabel('SPM {t}', fontsize=20)
    ax.set_ylim([-7, 7])
    plt.xticks(fontsize=17)
    plt.yticks(fontsize=17)
    plt.tight_layout()

    ### Incline
    # plot mean and SD:
    condition = "Incline"
    title_flat = f"{leg_position} - {load} {condition}"
    # ax = AX[0, 1]
    ax = fig.add_subplot(gs[0, 1])
    ax.text(-0.1, 1.1, "2a", transform=ax.transAxes, size=18, weight='bold')
    ax.set_title(title_flat, fontsize=24)
    plt.sca(ax)
    spm1d.plot.plot_mean_sd(exo_incline, linecolor='r', facecolor='r', label='Exo')
    spm1d.plot.plot_mean_sd(non_exo_incline, linecolor='b', facecolor='b', label='Non-Exo')
    ax.axhline(y=0, color='k', linestyle=':')
    # ax.set_xlabel('Stance phase (%)')
    ax.set_ylabel('Normalised GRF (BW)', fontsize=20)
    ax.set_xlabel('Stance phase (%)', fontsize=20)
    ax.set_ylim([0, 2.7])
    ax.set_xlim([0, 100])
    plt.xticks(fontsize=17)
    plt.yticks(fontsize=17)
    ax.legend(fontsize=17)

    # plot SPM results:
    # ax = AX[1, 1]
    ax = fig.add_subplot(gs[1, 1])
    ax.text(-0.1, 1.1, "2b", transform=ax.transAxes, size=18, weight='bold')
    plt.sca(ax)
    spmi2.plot()
    spmi2.plot_threshold_label(fontsize=17, pos=(20,5.5))
    spmi2.plot_p_values(size=10, offsets=[(0,0.9)])
    ax.set_xlabel('Stance phase (%)', fontsize=20)
    ax.set_ylabel('SPM {t}', fontsize=20)
    ax.set_ylim([-7, 7])
    plt.xticks(fontsize=17)
    plt.yticks(fontsize=17)
    plt.tight_layout()
    # plt.show()

    ### Decline
    # Plot mean and SD:
    condition = "Decline"
    title_flat = f"{leg_position} - {load} {condition}"
    # ax = AX[0, 2]
    ax = fig.add_subplot(gs[0, 2])
    ax.text(-0.1, 1.1, "3a", transform=ax.transAxes, size=18, weight='bold')
    ax.set_title(title_flat, fontsize=24)
    plt.sca(ax)
    spm1d.plot.plot_mean_sd(exo_decline, linecolor='r', facecolor='r', label='Exo')
    spm1d.plot.plot_mean_sd(non_exo_decline, linecolor='b', facecolor='b', label='Non-Exo')
    ax.axhline(y=0, color='k', linestyle=':')
    # ax.set_xlabel('Stance phase (%)')
    ax.set_ylabel('Normalised GRF (BW)', fontsize=20)
    ax.set_xlabel('Stance phase (%)', fontsize=20)
    ax.set_ylim([0, 2.7])
    ax.set_xlim([0, 100])
    plt.xticks(fontsize=17)
    plt.yticks(fontsize=17)
    ax.legend(fontsize=17)

    # plot SPM results:
    # ax = AX[1, 2]
    ax = fig.add_subplot(gs[1, 2])
    ax.text(-0.1, 1.1, "3b", transform=ax.transAxes, size=18, weight='bold')
    plt.sca(ax)
    spmi3.plot()
    spmi3.plot_threshold_label(fontsize=17, pos=(20,5.5))
    spmi3.plot_p_values(size=10, offsets=[(0,0.3)])
    ax.set_xlabel('Stance phase (%)', fontsize=20)
    ax.set_ylabel('SPM {t}', fontsize=20)
    ax.set_ylim([-7, 7])
    plt.xticks(fontsize=17)
    plt.yticks(fontsize=17)
    plt.tight_layout()

    # plt.savefig(f"{leg_position}{load}.png")

    return spmi1, spmi2, spmi3, fig
```
