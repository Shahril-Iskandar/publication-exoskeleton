```python
import pandas as pd
```

```python
# Read Excel file
bertec = pd.read_excel('Bertec_GRF_variables.xlsx')
loadsol = pd.read_excel('Loadsol_GRF variables.xlsx')
```

```python
loadsol
```

```python
# Pivot the DataFrame to transform the data
loadsol_impactforce = loadsol.pivot_table(index='ID', columns=['Exo', 'Condition', 'Load'], values=['Impact_Peak_Force_L', 'Impact_Peak_Force_R'])

# Flatten the multi-level columns
loadsol_impactforce.columns = ['_'.join(col).strip() for col in loadsol_impactforce.columns.values]

# Reset index to make 'ID' a column
loadsol_impactforce.reset_index(inplace=True)

# Save to Excel
loadsol_impactforce.to_csv("Loadsol_Impact_Force.csv", index=False)
```

```python
# Pivot the DataFrame to transform the data
loadsol_activeforce = loadsol.pivot_table(index='ID', columns=['Exo', 'Condition', 'Load'], values=['Active_Peak_Force_L', 'Active_Peak_Force_R'])

# Flatten the multi-level columns
loadsol_activeforce.columns = ['_'.join(col).strip() for col in loadsol_activeforce.columns.values]

# Reset index to make 'ID' a column
loadsol_activeforce.reset_index(inplace=True)

# Save to Excel
loadsol_activeforce.to_csv("Loadsol_Active_Force.csv", index=False)
```
